# 4.0 Diferença entre as rações 
Então a minha grande dúvida seria qual a diferença de componentes entre a Premier Filhotes Spitz e a Royal Canin?
Pois neste caso eu mudei da ração Premier Filhotes spitz e para a Royal Canin, e meu spitz melhorou do intestino. 
Então vamos lá. 

```python
df3=df2.copy()
```

Eu vou precisar dos nomes das colunas para identar e fazer a comparação. 

```python
lista = df3.columns
print(lista)
```

Como eu disse anteriormente eu tenho 5 marcar na minha base mas vou usar apenas 2. Então abaixo eu divido a base para utilizar apenas as rações que eu quero. 

```python
df4= df3[1:3]
```

```python
df4.tail()
```

Neste ponto vou verificar se um componente existe apenas em uma formula, e caso aconteça eu salvo o nome da coluna. 

```python
lista_de_componentes=[]
for i in lista:
    if df4[i][1]!= df4[i][2]:
        lista_de_componentes.append(i)
```

Agora abaixo é possivel verificar quais componentes são diferentes entre as duas rações, porém a lista abaixo possui elementos que existem na ração Premier e não existem na Royal, e componentes que não existem na Premier e existem na Royal. 

```python
lista_de_componentes
```

```python
df5=df4[lista_de_componentes]
```

```python
df5.tail()
```

Agora o proximo passo é selecionar apenas os componentes que existem na ração Premier e não existem na ração Royal. 

```python
lista_de_componentes_2=[]
for i in lista_de_componentes:
    if i != "Tipo de Racao":
        if df5[i][1]==1 and df5[i][2]==0:
            lista_de_componentes_2.append(i)
```

Pronto agora eu consegui verificar quais componentes podem ter desenvolvido alguma alergia no meu cãozinho.
